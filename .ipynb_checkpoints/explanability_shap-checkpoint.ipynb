```python
import lime
import pickle
import pandas as pd
import shap
pd.set_option('display.max_colwidth', None)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def load_data(path):
    with open(path, 'rb') as file:
        data = pickle.load(file)
        return data
```

# Explicabilité BA4

```python
base_path = 'Step3/explainability_results/shap/BA4'
```

```python
test_df = pd.read_csv(f'{base_path}/explain_data.csv', index_col=0)
```

```python
test_df
```

```python
X_test = test_df.iloc[:,:-3]
```

## Interpretabilité Globale

```python
explainer = load_data(f'{base_path}/explainer.pkl')
```

```python
shap_values = load_data(f'{base_path}/shap_values.pkl')
```

```python
exp_shap_values = load_data(f'{base_path}/exp_shap_values.pkl')
```

```python
shap.plots.bar(exp_shap_values, max_display=8)
```

```python
shap.plots.bar(exp_shap_values.cohorts(2).abs.mean(0), max_display=8)
```

```python
shap.plots.heatmap(exp_shap_values, max_display=8)
```

```python
shap.plots.beeswarm(exp_shap_values, max_display=8)
```

```python
shap.dependence_plot('mfcc3V_sma3nz_amean', shap_values, X_test)
```

```python
shap.dependence_plot('mfcc4V_sma3nz_amean', shap_values, X_test)
```

```python
shap.dependence_plot('mfcc2V_sma3nz_amean', shap_values, X_test)
```

## Interpretabilité Locale

```python
shap.plots.waterfall(exp_shap_values[0])
```

```python
shap.plots.waterfall(exp_shap_values[1])
```

```python
shap.plots.bar(exp_shap_values[0])
```

```python
shap.plots.bar(exp_shap_values[1])
```

```python
shap.decision_plot(explainer.expected_value, shap_values[0], X_test, link='logit', )
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
shap.decision_plot(explainer.expected_value, shap_values[1], X_test, link='logit', auto_size_plot=False)
```
